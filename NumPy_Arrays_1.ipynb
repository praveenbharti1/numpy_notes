### max,min,argmax,argmin

These are useful methods for finding max or min values.
Or to find their index locations using argmin or argmax

```python
arr = np.arange(1,23)
```

```python
arr
```

```python
arr.max() ## gives maximum value of element
```

```python
arr.argmax()  ## gives index location of maximum value
```

```python
arr.min() ##gives minimum value of element
```

```python
arr.argmin() ## gives location of minimum value.
```
